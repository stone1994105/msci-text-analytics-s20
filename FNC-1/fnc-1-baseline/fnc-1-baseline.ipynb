```python
def get_dataFrame(data):
    d = list()
    for stance in data.stances:
        d.append([stance['Headline'],stance['Body ID']])
    df = pd.DataFrame(d,columns = ['Headline','Body ID'])
    return df
```

```python
test_data = DataSet("competition_test")
test_df = get_dataFrame(test_data)
test_df
```
